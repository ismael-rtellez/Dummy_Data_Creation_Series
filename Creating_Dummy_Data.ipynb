## Problem 1: Creating random numbers

```python
import numpy as np

np.random.seed(123)       #we set seed value to avoid changes
num_aleatorios = np.random.multivariate_normal(mean=(-3,0), cov=([[1.0, 0.8], [0.8, 1.0]]), size= 500)  #generate the numbers using "np.random.multivariate_normal"
print(num_aleatorios)
```

## Problem 2: Visualization by scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np

x, y = np.hsplit(num_aleatorios, [1])   #split the matrix to have two columns of data of one dimension

#we plot and customize the graph according to our preference
fig, ax = plt.subplots()
plt.title("Scatter", size=20)
plt.xlabel("x1", size=14)
plt.ylabel("x2", size=14)
ax.scatter(x, y, color="red")
plt.grid()
plt.show()
```

## Problem 3: Visualization by histogram

```python
import matplotlib.pyplot as plt
import numpy as np

hist_1, hist_2 = np.hsplit(num_aleatorios, [1])   #split the matrix to have two columns of data of one dimension

#we plot and customize the graph according to our preference
fig, ax = plt.subplots()
ax.set_title("Histogram of x1", size=20)
ax.set_xlabel('x1', size=14)
ax.set_ylabel('Frequency', size=14)
ax.hist(hist_1, bins=50)
plt.xlim(-5.55, 2.7)
plt.show()

#we plot and customize the graph according to our preference
fig, ax = plt.subplots()
ax.set_title("Histogram of x2", size=20)
ax.set_xlabel('x2', size=14)
ax.set_ylabel('Frequency', size=14)
ax.hist(hist_2, bins=50)
plt.xlim(-5.55, 2.7)
plt.show()
```

## Problem 4: Addition of Data

```python
import numpy as np

np.random.seed(321)        #we set seed value to avoid changes

#generate the numbers using "np.random.multivariate_normal"
num_aleatorios_2 = np.random.multivariate_normal(mean=(0, -3), cov=([[1.0, 0.8], [0.8, 1.0]]), size= 500)

#we plot both matrix and customize the graph according to our preference
x, y = np.hsplit(num_aleatorios, [1])
x2, y2 = np.hsplit(num_aleatorios_2, [1])
fig, ax = plt.subplots()
plt.title("Scatter", size=20)
plt.xlabel("x1", size=14)
plt.ylabel("x2", size=14)
ax.scatter(x, y, color= "red")
ax.scatter(x2, y2, color="blue")
ax.legend(["0", "1"])   #Add the legend to identify each matrix
plt.grid()
plt.show()
```

## Problem 5: Data combination

```python
import numpy as np

matrix_mixed = np.vstack([num_aleatorios, num_aleatorios_2]) #Use "vstack" for connect the two matrixs
print(matrix_mixed)
np.shape(matrix_mixed)  #Verify the shape for see that the two matrixs are connected
```
